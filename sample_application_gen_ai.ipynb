```python
# Step 2: Set API Key Securely
import os
import getpass
import google.generativeai as genai

# Prompt for your API key securely
api_key = getpass.getpass("ğŸ”‘ Enter your Gemini API Key: ")
os.environ["GOOGLE_API_KEY"] = api_key

# Configure Gemini
genai.configure(api_key=api_key)
```

```python
# Step 3: Initialize Chat Model (with Memory)

# Create a Gemini chat model with memory (context)
model = genai.GenerativeModel("gemini-1.5-pro", generation_config={"temperature": 0.9} )

chat = model.start_chat(history=[])
```

```python
# Step 4: Chat Function (Text Only)
from IPython.display import Markdown

def chat_with_gemini(user_input):
    try:
        response = chat.send_message(user_input)
        return response.text
    except Exception as e:
        return f"âŒ Error: {e}"
```

```python
# Step 5: Interactive Chat Loop
print("ğŸ’¬ Gemini is ready! Type 'exit' to quit.\n")

while True:
    user_input = input("You: ")
    if user_input.strip().lower() == "exit":
        print("ğŸ‘‹ Goodbye!")
        break
    response = chat_with_gemini(user_input)
    display(Markdown(f"**Gemini:** {response}"))
```

```python
!pip install google-generativeai pymupdf


import google.generativeai as genai
from google.colab import files
from PIL import Image, UnidentifiedImageError
from IPython.display import Markdown
import fitz  # PyMuPDF for PDF
import io

# Configure Gemini model
genai.configure(api_key="YOUR_API_KEY")  # Replace with your actual API key
model = genai.GenerativeModel(model_name="gemini-pro")  # RAG works with text-based input

# Upload file
uploaded = files.upload()
file_name = list(uploaded.keys())[0]
file_data = uploaded[file_name]
file_ext = file_name.split('.')[-1].lower()

# Extract context from PDF
pdf_text = None
if file_ext == "pdf":
    with fitz.open(stream=file_data, filetype="pdf") as doc:
        pdf_text = ""
        for page in doc:
            pdf_text += page.get_text()
        pdf_text = pdf_text[:3000]  # Limit length to avoid hitting token limits

# Prompt
user_question = input("Ask a question based on the uploaded file: ")

# RAG-style prompt
if pdf_text:
    context = f"You are an expert assistant. Use the following document as context:\n\n{pdf_text}\n\n"
    full_prompt = context + f"Now, answer this question: {user_question}"
    response = model.generate_content(full_prompt)
    display(Markdown(f"**Answer:** {response.text}"))
else:
    print("Only PDF is supported in this version. Image input requires a vision model.")
```

```python
# Step 6: View Conversation History (Optional)

print("ğŸ“ Conversation history:")
for i, msg in enumerate(chat.history):
    role = "You" if msg.role == "user" else "Gemini"
    print(f"{role}: {msg.parts[0].text}\n")
```
